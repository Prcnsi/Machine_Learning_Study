```python
# 교차 검증
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris=load_iris()
features=iris.data
label=iris.target
D_Tree=DecisionTreeClassifier(random_state=156)

#데이터 5개로 분리 !!
kfold=KFold(n_splits=5)
cv_accuracy=[]
print('붓꽃 데이터 Set Size:',features.shape[0])
```

```python
# 대충 5번에 나눠서 학습시키는 코드!
n=0
for train_index,test_index in kfold.split(features):
    X_train,X_test=features[train_index],features[test_index]
    y_train, y_test=label[train_index],label[test_index]
    
    D_Tree.fit(X_train,y_train)
    pred=D_Tree.predict(X_test)
    n+=1
    
    accuracy=np.round(accuracy_score(y_test,pred),4)
    train_size=X_train.shape[0]
    test_size=X_test.shape[0]
    print('\n*{0} 번째 검증 정확도:{1}'.format(n,accuracy))
    print('\n*{0} 번째 Set의 index:{1}\n\n'.format(n,test_index))
    cv_accuracy.append(accuracy)
    
print('\n # 평균 검증 정확도:',np.mean(cv_accuracy))
```

```python
#Stratified K 폴드 -> 불균형적인 분포를 개선
#+ 불균형한 분포도를 가진 레이블을 위한 방식
import pandas as pd
iris=load_iris()
iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts()
```

```python
# n_splits(n): n개로 분할!
kfold=KFold(n_splits=3)
n=0
for train_index,test_index in kfold.split(iris_df):
    n+=1
    label_train=iris_df['label'].iloc[train_index]
    label_test=iris_df['label'].iloc[test_index]
    print('*교차 검증: {0}'.format(n))
    print('\n학습 데이터 분포:\n',label_train.value_counts())
    print('\n테스트 데이터 분포:\n',label_test.value_counts())
```
