# 데이터 전처리
* 참고 영상: https://www.youtube.com/watch?v=3-iwY0_u0VA
* 느낀점: 데이터 전처리는 그냥 raw data를 보기쉽게 필요한 것만 남겨서 재정리 하는 것이구나.  
그러면 여러 데이터를 갖다가 문제 풀듯이 연습해야겠다.

```python
import pandas as pd
```

```python
# skiprows=n 은 위에서부터 n개의 행 생략하고 출력
data=pd.read_excel('C:/Users/J-365/Downloads/Data05.xlsx',skiprows=1)

# print(data)보다 data 명령이 더 깔끔하게 데이터 출력함
data
```

```python
# 데이터 구조 확인 shape 은 (행,열) 구조로 반환
data.shape
```

```python
# 칼럼 목록들 출력
data.columns
```

```python
# iloc[index, column] = 전체 index행에서 column 로우를 가져옴, 유사 좌표 !! [행,열]
# 여기서는 비교적 불필요한 정보들만 출력해 본 것
data.iloc[0:3,8:]
```

```python
data.iloc[:,8:-2]
```

```python
# 위에서 날짜 위치를 바꾸기 위해 행과 열을 바꿔주는 T를 사용
# T는 행, 열의 목록을 바꿈
# 기준이 되는 T값을 잘 설정해 주어야 함!
data.iloc[:,8:-2].T
```

```python
# 시리즈 형태로 쌓아서 정리
stack_data=data.iloc[:,8:-2]
stack_data.stack()
```

```python
# 데이터 프레임으로 변환
pd.DataFrame(stack_data.stack())
```

```python
# 인덱스 리셋
pd.DataFrame(stack_data.stack()).reset_index()
```

```python
# 정렬한 데이터에서 날자별로 추출해서 재고양 계산 가능
pd.DataFrame(stack_data.stack()).reset_index().iloc[1:]
```

```python
# T값을 다시 잡고 재정렬
data.iloc[:,6:-2]
```

```python
# 불필요한 칼럼 제외
new_data=data.iloc[:6:-2].drop(columns=['Unnamed: 7','판매'])
```

```python
new_data
```

```python
new_data.head(2)
```

```python
# 제품명을 칼럼명으로 지정
new_data.set_index(['제품명']).stack
```

```python
# 데이터 프레임으로 변환
pd.DataFrame(new_data.set_index(['제품명'])).stack().reset_index()
```
